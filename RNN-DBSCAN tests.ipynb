# RNN-DBSCAN clustering tests
The original author's did not release their code, so these tests are for me to see if my implementation is correct, or at least performs reasonably.

# Tests on datasets used in their paper
They report performance on some shape-based synthetic datasets (Table 3). I am going to attempt to repeat their results on these datasets:
- aggregation
- d31
- flame
- jain
- r15
- spiral
- pathbased

These datasets are from the [University of Finland shape sets](http://cs.uef.fi/sipu/datasets/). Some of them are directly from that site, and some are from this [Github repo](https://github.com/milaan9/Clustering-Datasets) that converted the datasets into mat files. 

In the paper, they report the Adjusted Rand Index (ari), purity, # of clusters, and # of noise points. I am just going to compare ari, # of clusters, and # of noise points.

They sweep $k$ from 1 to 100 and find the $k$ that results in the max ARI, so I will do the same.

`rand_index` is by Chris McComb on [GitHub](https://github.com/cmccomb/rand_index); it needs to be on the MATLAB path

## Aggregation

### original labels

```matlab
load([dataDir filesep 'aggregation.mat'])
scatter(data(:,1), data(:,2), 40, label, 'filled')
```

### RNN-DBSCAN results

```matlab
maxAri = intmin;
kMaxAri = 1;

for k = 1:100
    rnnDbscan = RnnDbscan(data, k);
    rnnDbscan.cluster();
    ari = rand_index(label, rnnDbscan.Labels, 'adjusted');
    if ari > maxAri
        maxAri = ari;
        kMaxAri = k;
    end
end
```

```matlab
rnnDbscan = RnnDbscan(data, kMaxAri);
rnnDbscan.cluster();
```

```matlab
scatter(rnnDbscan.Data(:,1), rnnDbscan.Data(:,2), 40, rnnDbscan.Labels, 'filled')
```

```matlab
maxAri
length(rnnDbscan.Clusters)
length(rnnDbscan.Outliers)
```

They were able to achieve a max ari of 0.998 and had 0 noise points. My results are not quite that good, but they are still close. There might be something wrong in my algorithm that is improperly assigning noise points

## Thoughts
The results on these synthetic datasets do not perfectly match the results reported in the paper, but they are "close". 
There could be an issue in my implementation. However, they do mention that non-core observations may be reachable from multiple clusters, and that such observations much be assigned to only one cluster in a hard clustering; their implementation "takes an arbitrary approach to the cluster assignment of such border observations that is dependent on observeration ordering". So it could be the case that I am assigning border observations in a different manner than they do, resulting in different results.

I am pretty confident that my implementation follows the graph-based interpretation of the algorithm given in the paper.

## Closing thoughts
Overall, my implementation seems to perform pretty well on most of these datasets. However, the results do seem pretty sensitive to $k$. When I was just manually choosing values for $k$ instead of sweeping $1 \le k \le 100$, the results varied quite a bit. Using the heuristics presented in the paper for choosing $k$ do seem like they will work fine in practice, though.
